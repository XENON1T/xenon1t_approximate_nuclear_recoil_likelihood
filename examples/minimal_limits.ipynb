# Minimal example
In this low-weight example, we will provide an energy spectrum in its native form - python callable:

```python
import scipy.stats as sps
spectrum = sps.norm(10, 0.1).pdf
```

Then we create an inference object from XENONnT toyMC simulations and the spectrum

```python
from xe_likelihood import BinwiseInference
xenon_inference = BinwiseInference.from_xenonnt_mc(spectrum=spectrum)
```

Then we compute the upper limits on all simulations and show their percentiles

```python
uls = xenon_inference.compute_uls(cl=0.1)
xenon_inference.compute_ul_percentiles()
```

we can also plot the distribution of expected limits under the null hypothesis|

```python
xenon_inference.plot_histogram()
```

```python
xenon_inference.plot_migration_matrix(show=True)
```
