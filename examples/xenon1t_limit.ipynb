## Estimating the XENON1T limit for a 50 GeV WIMP

```python
from xe_likelihood import BinwiseInference, Spectrum
```

#### Step 1: define the recoil spectrum
In this case we will use the built in WIMP spectrum

```python
spectrum_wimp = Spectrum.from_wimp(mass=50)
```

#### Step 2: Create an inference object from the XENON1T results
These results are distributed with the python package

```python
xenon_inference = BinwiseInference.from_xenon1t_sr(spectrum=spectrum_wimp)
```

#### Step 3: Make some plots!

```python
xenon_inference.plot_spectrum(show=True)
xenon_inference.plot_migration_matrix(show=True)
xenon_inference.plot_likelihood_matrix(show=True)
xenon_inference.compute_ul(spectrum=spectrum_wimp)
print(xenon_inference.uls)
```
