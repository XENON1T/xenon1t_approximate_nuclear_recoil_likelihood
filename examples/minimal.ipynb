# Minimal example

In this low-weight example, we will provide an energy spectrum in its native form - python callable and make a 90% confidence level limit using the XENON1T data:

```python
import scipy.stats as sps
from xe_likelihood import BinwiseInference

xenon_inference = BinwiseInference.from_xenon1t_sr(spectrum = sps.norm(10, 0.1).pdf)
asymptotic_limit = xenon_inference.compute_ul(cl=0.1, asymptotic=True)
nonasymptotic_limit = xenon_inference.compute_ul(cl=0.1, asymptotic=False)
print ("Asymptotic Limit =", asymptotic_limit)
print ("Non-asymptotic Limit =", nonasymptotic_limit)
```
